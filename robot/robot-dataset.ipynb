```python
import pandas as pd
df = pd.read_excel('lp1.xlsx')
df.head(20)
```

```python
# delete empty rows
df = df.dropna(how='all')
df.head(20)
```

```python
#fill empty celles in class column based on the previous value
df['class'] = df['class'].fillna(method='ffill')
df.head(20)
```

```python
# delete rows with missing values
df = df.dropna()
```

```python
# apply ordinal encoder to the class column 
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df['class-oridnal'] = encoder.fit_transform(df[['class']])
df.head(20)
```

```python
df.to_excel('lp1_cleaned.xlsx', index=False)
```
